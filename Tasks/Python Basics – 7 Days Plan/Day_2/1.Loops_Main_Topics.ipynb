###Iterating over Lists

````python
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
````

###Iterating over Strings

````python
name = "Mohan"
for char in name:
    print(char)
````

###Iterating over Tuples

###Iterating with Index

````python
colors = ['red', 'green', 'blue']
for index, color in enumerate(colors):
    print(index, color)
````

#2. while Loop in Python
####A while loop keeps running as long as the condition is True.

````python
#Print numbers from 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1
````

````python
# Count down from 5 to 1
i = 5
while i >= 1:
    print(i)
    i -= 1
````

````python
# Using while with break (exit loop early)
i = 1
while i <= 10:
    if i == 8:
        break
    print(i)
    i += 1
````

````python
#Using while with continue (skip current iteration)
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
````

````python
#checking correct password
password = ""
while password != "admin123":
    password = input("Enter password: ")
print("Access granted!")
````

#3. Loop Control Statements in Python


###Loop control statements are used to change the flow of execution inside loops.



````python
##break – Stop the loop immediately
for i in range(1, 10):
    if i == 5:
        break
    print(i)
````

````python
#continue – Skip current iteration and continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
````

````python
#pass – Do nothing (placeholder)
for i in range(1, 6):
    if i == 3:
        pass  # do nothing
    print(i)
````

#Loop with else in Python


````python
#for loop with else
for i in range(1, 6):
    print(i)
    if i == 3:
        break
else:
    print("Loop completed successfully.")
````

````python
#for loop with else (no break)
for i in range(1, 4):
    print(i)
else:
    print("Loop completed without break.")
````

````python
#Prime Number Check using for-else
num = int(input("Enter a number: "))

for i in range(2, num):
    if num % i == 0:
        print(f"{num} is not prime.")
        break
else:
    print(f"{num} is a prime number.")
````

````python
#while loop with else
i = 1
while i <= 3:
    print(i)
    i += 1
else:
    print("While loop ended normally.")
````

# Nested Loops in Python

````python
#Print a 3x3 matrix
for i in range(3):
    for j in range(3):
        print("*", end=" ")
    print()
````

````python
#Print a Right-Angled Triangle Pattern
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
````

````python
#Print a Number Pattern
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
````

````python
#Multiplication Table (1 to 5)
for i in range(1, 6):
    for j in range(1, 6):
        print(i * j, end="\t")
    print()
````
